# Fruit - Preliminary Examination


```python
%matplotlib notebook
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# fruits dataset
fruits = pd.read_table('data/fruit_data_with_colors.txt')
```

```python
# use the 'seaborn-colorblind' style
#plt.style.use('seaborn-colorblind')
plt.style.use('bmh')
```

```python
sns.pairplot(fruits, vars=['mass', 'width', 'height', 'color_score'], hue='fruit_name', diag_kind='kde');
```

```python
plt.figure(figsize=(8,6))
for i, feature in enumerate(['mass', 'width', 'height', 'color_score']):
    plt.subplot(2, 2, i+1)
    sns.violinplot('fruit_name', feature, data=fruits)
```
